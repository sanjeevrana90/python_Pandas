### Loading Data into Pandas


```python
import pandas as pd

poke = pd.read_csv('pokemon_data.csv')
print(poke.head(5))
```

```python
print(poke.tail(3))
```

### Reading Data in Pandas

```python
## Read headers

print(poke.columns)
```

```python
## Read each Column

print(poke['Name'][:5])
print(poke[['Name', 'Type 1', 'HP']])
```

```python
## Read Each row - iloc = indent location

print(poke.iloc[1:4])
print(poke.iloc[4])
```

```python
## Read a specific Location (R, C)

print(poke.iloc[3,1])
```

```python
## Locating pokemons where Type 1 = fire

poke.loc[poke['Type 1'] == "Fire"]
```

## Sorting / Describing Data

```python
poke.describe()
```

```python
poke.sort_values(['Type 1', 'HP'])
```

```python
poke.sort_values(['Type 1', 'HP'], ascending = False)

poke.sort_values(['Type 1', 'HP'], ascending = [1,0])
```

## Making changes to the Data

```python
poke['Total'] = poke['HP'] + poke['Attack'] + poke['Defense'] + poke['Sp. Atk'] + poke['Sp. Def'] + poke['Speed']

poke.head(5)
```

```python
# Adding columns using index number

poke['Total'] = poke.iloc[:, 4:10].sum(axis = 1)

poke.head(5)
```

```python
# Reorganizaing the columns 

cols = list(poke.columns)
poke = poke [cols[0:4] + [cols[-1]] + cols [4:12]]

poke.head(5)
```

### Filtering Data

```python
poke.loc[(poke['Type 1'] == 'Grass') & (poke['Type 2'] == 'Poison') & (poke['HP'] > 70)]
```

```python
# exporting Filtered data

new_poke = poke.loc[(poke['Type 1'] == 'Grass') & (poke['Type 2'] == 'Poison') & (poke['HP'] > 70)]

new_poke.to_csv('filtered.csv')
# new_poke.to_csv('filtered.txt', index = False, sep = '\t')
```

```python
# Filtering out pokemons names whose name contains 'Mega'

poke.loc[poke['Name'].str.contains('Mega')]
```

```python
# Filtering out pokemons names whose name does not contain 'Mega'

poke.loc[~poke['Name'].str.contains('Mega')]
```

```python
# filtering with regular expressions with ignoring case sensitiveness 

import re

poke.loc[poke['Type 1'].str.contains('fire|grass', flags = re.I, regex = True)]
```

## Aggregate Statistics

```python
poke.groupby(['Type 1']).mean().sort_values('Attack', ascending = False)

poke.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)
```

```python
poke.groupby(['Type 1']).sum()
```

```python
poke['count'] = 1

poke.groupby(['Type 1']).count()['count']
```
